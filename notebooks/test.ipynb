```python
import boost_histogram as bh
import mplhep as hep
import matplotlib.pyplot as plt
plt.style.use(hep.style.ROOT)

nbin, xmin, xmax = 40, -10, 10

hist = bh.Histogram(bh.axis.Regular(nbin, xmin, xmax), storage=bh.storage.Weight())
# hist.fill(df.npfcand[df.pass_fiducial])
# hist.fill(ak.flatten(df[df.pass_fiducial][:100].pfcand_ptrel))
hist.fill(ak.flatten(df[df.pass_fiducial][:100].pfcand_e_reg))
content, yerr = hist.view().value, np.sqrt(hist.view().variance)

f, ax = plt.subplots(figsize=(10,10))
hep.histplot(content, bins=hist.axes[0].edges, yerr=yerr)
ax.set_xlabel('Label', ha='right', x=1.0); ax.set_ylabel('Events / bins', ha='right', y=1.0)
```

```python
import re
r = re.compile('QCD_Pt_(\w*)_')
lines = open('test.txt').readlines()
```

```python
d = {}
for i, l in enumerate(lines):
    s = re.findall(r, l)[0]
    if s not in d:
        d[s] = []
    d[s].append(i)
```

```python
rm_list = []
for k in d:
    rm_list.append(d[k][100:])
rm_list = [e for sublist in rm_list for e in sublist]
```

```python
with open('rm_list.txt', 'w') as f:
    f.write('\n'.join([str(e) for e in rm_list]))
```
